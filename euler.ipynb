```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función de la ecuación diferencial
def f(t, y):
    return y - t**2 + 1

# Implementación del método de Euler
def ODE_euler(a, b, f, y_t0, N):
    h = (b - a) / N
    t = a
    ts = [t]
    ys = [y_t0]

    for _ in range(N):
        y = ys[-1]
        y += h * f(t, y)
        ys.append(y)

        t += h
        ts.append(t)
    return ys, ts

# Valores iniciales
a = 0.0
b = 2.0
y_t0 = 0.5

# Valores de N
N_values = [5, 10, 20]

# Graficar soluciones
plt.figure(figsize=(8, 5))

for N in N_values:
    ys, ts = ODE_euler(a, b, f, y_t0, N)
    plt.plot(ts, ys, marker='o', label=f'N={N}')

plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Método de Euler para dy/dt = y - t**2 + 1')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de la ecuación diferencial
def f(t, y):
    return y - t**2 + 1

# Implementación del método de Euler
def ODE_euler(a, b, f, y_t0, N):
    h = (b - a) / N
    t = a
    ts = [t]
    ys = [y_t0]

    for _ in range(N):
        y = ys[-1]
        y += h * f(t, y)
        ys.append(y)

        t += h
        ts.append(t)
    return ys, ts

# Definir la solución exacta
def exact_solution(t):
    return (t + 1)**2 - 0.5 * np.exp(t)

# Parámetros
a = 0.0
b = 2.0
y_t0 = 0.5
N = 10

# Obtener la solución numérica con Euler
ys_euler, ts_euler = ODE_euler(a, b, f, y_t0, N)

# Calcular la solución exacta en los mismos puntos
ys_exact = [exact_solution(t) for t in ts_euler]

# Graficar la solución exacta y la aproximada
plt.figure(figsize=(8, 5))

plt.plot(ts_euler, ys_euler, marker='o', label='Euler (N=10)', linestyle='--')
plt.plot(ts_euler, ys_exact, label='Solución exacta', linewidth=2)

plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Método de Euler vs. Solución Exacta')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de la ecuación diferencial
def f(t, y):
    return y - t**2 + 1

# Implementación del método de Euler
def ODE_euler(a, b, f, y_t0, N):
    h = (b - a) / N
    t = a
    ts = [t]
    ys = [y_t0]

    for _ in range(N):
        y = ys[-1]
        y += h * f(t, y)
        ys.append(y)

        t += h
        ts.append(t)
    return ys, ts

# Definir la solución exacta
def exact_solution(t):
    return (t + 1)**2 - 0.5 * np.exp(t)

# Parámetros
a = 0.0
b = 2.0
y_t0 = 0.5
N_values = [5, 10, 20]

# Crear la gráfica
plt.figure(figsize=(8, 5))

# Graficar la solución exacta en un rango continuo
t_exact = np.linspace(a, b, 100)
y_exact = exact_solution(t_exact)
plt.plot(t_exact, y_exact, label="Solución exacta", linewidth=2, color='black')

# Calcular y graficar la solución de Euler para cada N
for N in N_values:
    ys_euler, ts_euler = ODE_euler(a, b, f, y_t0, N)
    plt.plot(ts_euler, ys_euler, marker='o', linestyle='--', label=f'Euler (N={N})')

plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Método de Euler vs. Solución Exacta')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de la ecuación diferencial
def f_new(t, y):
    return (y / t) - (y / t)**2

# Implementación del método de Euler
def ODE_euler(a, b, f, y_t0, N):
    h = (b - a) / N
    t = a
    ts = [t]
    ys = [y_t0]

    for _ in range(N):
        y = ys[-1]
        y += h * f(t, y)
        ys.append(y)

        t += h
        ts.append(t)
    return ys, ts

# Definir la solución exacta
def exact_solution_new(t):
    return t / (1 + np.log(t))

# Parámetros
a_new = 1.0
b_new = 2.0
y_t1 = 1.0
N_values = [5, 10, 20]

# Crear la gráfica
plt.figure(figsize=(8, 5))

# Graficar la solución exacta en un rango continuo
t_exact_new = np.linspace(a_new, b_new, 100)
y_exact_new = exact_solution_new(t_exact_new)
plt.plot(t_exact_new, y_exact_new, label="Solución exacta", linewidth=2, color='black')

# Calcular y graficar la solución de Euler para cada N
for N in N_values:
    ys_euler, ts_euler = ODE_euler(a_new, b_new, f_new, y_t1, N)
    plt.plot(ts_euler, ys_euler, marker='o', linestyle='--', label=f'Euler (N={N})')

plt.xlabel('t')
plt.ylabel('y(t)')
plt.title("Método de Euler vs. Solución Exacta")
plt.legend()
plt.grid()
plt.show()
```
